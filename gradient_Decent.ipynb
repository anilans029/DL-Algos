```python
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80,random_state=13)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = "red")
```

```python
# apply gradient descent assuming slope is constant m = 78.5 and b = 0
y_pred=((78.5*x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color = "red",label = "ols")
plt.plot(x,y_pred,color = "cyan",label = "b = 0")
plt.legend()
```

```python
epochs = 10
m = 78.35
b  =100
lr = 0.01
y_pred = m*x+b
# plt.plot(x,y_pred)
for i in range(epochs):
    slope_b = -2*np.sum(y-m*x.ravel() - b)
    slope_m = -2*x.ravel()*np.sum(y-m*x.ravel() - b)
    
    b = b - lr*slope_b
    m = m - lr*slope_m
    
    y_pred = m*x+b
    plt.plot(x,y_pred)
print("_intercept : ",b)
print("_coeff : ",m)
plt.scatter(x,y)
```
